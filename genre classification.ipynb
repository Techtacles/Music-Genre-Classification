### USING ARTIFICIAL NEURAL NETWORK FOR GENRE CLASSIFICATION

```python
import numpy as np
import pandas as pd
import librosa
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import json
import librosa.display
from sklearn.model_selection import train_test_split
import tensorflow as tf
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

Be sure to use the correct file paths in your directory

```python
dataset_path=r"Music genre classification\genres_original"
```

```python
data={"labels":[],"audio":[],"mappings":[]}
#emp=[]
def preprocess(dataset_path,n_mfcc=13,hop_length=512,n_fft=2048):
    for folder,dirname,filename in os.walk(dataset_path):
        if dirname!=[]:
            data["labels"].append(dirname)
            for i in data["labels"]:
                for j in i:
                    pa=os.path.join(dataset_path,j)
                    #print(pa)
                    for k in os.listdir(pa):
                        #print(k)
                        path=os.path.join(pa,k)
                        pp=path.split("\\")
                        #print(path)
                        audio,sr=librosa.load(path)
                        mfcc=np.mean(librosa.feature.mfcc(audio,sr=sr,n_mfcc=n_mfcc,hop_length=hop_length,n_fft=n_fft).T,axis=0)
                        data["audio"].append(mfcc)
                        data["mappings"].append(pp[-2])
                        #emp.append(mfcc)
```

```python
preprocess(dataset_path)
```

```python
data["audio"]
```

```python
def turn_into_inputs_and_targets(x):
    inputs=np.array(x["audio"])
    targets=np.array(x["mappings"])
    return inputs,targets
```

```python
inputs,targets=turn_into_inputs_and_targets(data)
```

```python
targets
```

Encoding the categorical features

```python
le=LabelEncoder()
```

```python
encoded_targets=le.fit_transform(targets)
```

```python
encoded_targets
```

```python
inputs
```

##### SPLITTING THE DATA

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(inputs,encoded_targets,test_size=0.25)
```

```python
Xtrain.shape,ytrain.shape
```

```python
Xtest.shape,ytest.shape
```

##### BUILDING ARTIFICIAL NEURAL NETWORK ARCHITECTURE

```python
from tensorflow.keras.layers import Dense,Flatten,Dropout,BatchNormalization,Conv2D,MaxPool2D
from tensorflow.keras.models import Sequential
```

```python
model=Sequential([
    Flatten(input_shape=(13,)),
    Dense(100,activation="relu"),
    Dropout(0.2),
    BatchNormalization(),
    Dense(100,activation="relu"),
    Dropout(0.2),
    BatchNormalization(),
    Dense(100,activation="relu"),
    Dropout(0.2),
    BatchNormalization(),
    #Dense(100,activation="relu"),
       
    #output layer
    Dense(10,activation="softmax")
    
])

model.compile(loss="sparse_categorical_crossentropy",optimizer="rmsprop",metrics=["accuracy"])
model.fit(Xtrain,ytrain,validation_data=(Xtest,ytest),epochs=100)
model.evaluate(Xtest,ytest)
```

### PREDICTIONS AND TESTING

```python
def predict(path):
    audio,sr=librosa.load(path)
    mfcc=np.mean(librosa.feature.mfcc(audio,sr=sr,n_mfcc=13,hop_length=512,n_fft=2048).T,axis=0)
    pred=model.predict(mfcc.reshape(1,13))
    pred_value=np.argmax(pred)
    converted=le.inverse_transform([pred_value])
    return converted
```

Now let us test our prediction with a file in the hippop directory

```python
predict(r"Music genre classification\genres_original\hiphop\hiphop.00004.wav")
```
